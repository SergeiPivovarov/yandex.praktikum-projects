### Retantion Rate

Для ответа на вопрос, как часто возвращаются люди, рассчитаем метрику Retantion Rate и проведем когортный анализ. Результаты отобразим на тепловой карте.

Определяем дату первого визита пользователей.

```python
first_visit = visits.groupby(['uid'])['date'].min()
```

```python
first_visit.name = 'first_visit'
```

Добавляем её к основной таблице с визитами.

```python
visits = visits.join(first_visit, on='uid')
```

Считаем lifetime когорт.

```python
visits['cohort_lifetime'] = visits['date'] - visits['first_visit']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1, 'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
```

Добавляем столбец с месяцем первого визита.

```python
visits['first_visit_month'] = visits['first_visit'].astype('datetime64[M]')
```

Считаем размеры когорт.

```python
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index() 
```

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]
```

```python
initial_users_count.sample(5)
```

```python
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
```

Добавляем размеры когорт в таблицу.

```python
cohorts = cohorts.merge(initial_users_count, on='first_visit_month')
```

Считаем retention rate.

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
```

```python
cohorts.sample(5)
```

Улучшаем вид даты.

```python
cohorts['first_visit_month'] = cohorts['first_visit_month'].dt.strftime('%Y-%m')
```

Строим сводную таблицу когортного анализа и визуализируем.

```python
retention_pivot = cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='mean'
)
```

```python
plt.figure(figsize=(13,9))
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmin=0.003, vmax=0.074)
plt.title('Когортный анализ - retention rate')
plt.xlabel('Lifetime')
plt.ylabel('Когорты')
```

- После знакомства с сервисом в последующем в него возвращаются менее 8% пользователей.
- С увеличением времени жизни по всем когортам прослеживается динамика уменьшения количества пользователей, retention rate уменьшается.
- Чем моложе когорта, тем меньше retention rate второго месяца жизни, то есть удерживать пользователей впервые зашедших на сервис, становится все сложнее.

В целом, учитывая специфику сервиса, результаты не являются неоджиданными. Субъективно предполагаю, что основная масса людей посещают культурные мероприятия только несколько раз в год. Доля людей, регулярно посещающих мероприятия, невелика, поэтому и возвращается после первой покупки лишь небольшой процент клиентов.

### Подготовка таблицы заказов к анализу

Подготавливаем таблицу заказов к анализу всех остальных метрик: находим первый день заказа, первый месяц заказа, считаем lifetime каждого заказа.

```python
orders.sample()
```

Выделяем день из столбца с датой и временем заказа.

```python
orders['date'] = orders['buy_ts'].astype('datetime64[D]')
```

```python
first_order = orders.groupby(['uid'])['date'].min()
```

```python
first_order.name = 'first_order'
```

Добавляем первый день заказа к исходной таблице.

```python
orders = orders.join(first_order, on='uid')
```

Считаем lifetime когорт.

```python
orders['cohort_lifetime'] = orders['date'] - orders['first_order']
orders['cohort_lifetime'] = orders['cohort_lifetime'] / np.timedelta64(1, 'M')
orders['cohort_lifetime'] = orders['cohort_lifetime'].round().astype('int')
```

### Количество покупок на пользователя

Считаем количество заказов по месяцу первого заказа и lifetime.

```python
orders_per_user = orders.groupby(['first_order_month', 'cohort_lifetime'])['uid'].count().reset_index()
```

```python
orders_per_user.columns = ['first_order_month', 'cohort_lifetime', 'orders_count']
```

Считаем размеры когорт по первому месяцу заказа.

```python
users_count = orders.groupby(['first_order_month'])['uid'].nunique()
```

```python
users_count.name = 'cohort_users'
```

Добавляем размеры когорт в таблицу.

```python
orders_per_user = orders_per_user.merge(users_count, on='first_order_month')
```

Считаем метрику - количество заказов на пользователя.

```python
orders_per_user['orders_per_user'] = orders_per_user['orders_count'] / orders_per_user['cohort_users']
```

Улучшаем вид даты.

```python
orders_per_user['first_order_month'] = orders_per_user['first_order_month'].dt.strftime('%Y-%m')
```

Строим сводную таблицу когортного анализа и визуализируем.

```python
orders_per_user_pivot = orders_per_user.pivot_table(
    index='first_order_month',
    columns='cohort_lifetime',
    values='orders_per_user',
    aggfunc='mean'
)
```

## Источники

- **Источник 1.** Наиболее ценный источник. Имеет наибольший LTV и приемлемый САС. Расходы по нему окупаются уже на второй месяц.

  **Рекомендация:** Если возможно - масштабировать + увеличить бюджет, если нет - оставить все, как есть.
  
  
- **Источник 2.** Ценный источник. Высокий САС, но имет один из наибольших LTV (2 место). Расходы окупаютя на 5-6 месяц. 
  
  **Рекомендация:** Снизить САС + масштабировать + увеличить бюджет.


- **Источник 3.** Имеет наибольшую долю в расходах маркетинг и наибольший САС. LTV растет медленно (6 место) и уступает большинству других источников. Отстуствие существенной положительной динамики по ROI в когортном анализе говорит о том, что расходы по нему не окупятся.
  
  **Рекомендация:** Сильно снизить САС + уменьшить бюджет.


- **Источник 4.** Имеет приемлемый САС, хороший LTV (4 место). Расходы по нему окупаются медленно - на 10-ый месяц.
  
  **Рекомендация:** Снизить САС + бюджет оставить на текущем уровне.


- **Источник 5.** Имеет приемлемый САС, хороший LTV (3 место). Расходы по нему окупаются на 4-ый месяц.
  
  **Рекомендация:** Масштабировать + увеличить бюджет.


- **Источник 9.** Имеет приемлемый САС, хороший LTV (5 место). Расходы по нему окупаются на 4-5-ый месяц.
  
  **Рекомендация:** Масштабировать + увеличить бюджет.


- **Источник 10.** Имеет приемлемый САС, но наименьший LTV (7 место). Расходы по нему либо не окупятся, либо будут окупаться очень долго.
  **Рекомендация:** Уменьшить бюджет.

Общая рекомендация - увеличивать LTV.

Отключать никакие источники не стал бы, несмотря на плохие результаты. Лучше постепенно вносить изменения и отслеживать результаты. Есть понятия моделей атрибуции и ассоциированных конверсий, которы здесь никак не учитываются, поэтому к результатам анализа нужно относиться аккуратно - учитывать, но не считать полностью исчерпывающими.
